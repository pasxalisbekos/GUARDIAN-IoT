```python
import os
import json
import seaborn as sns
import matplotlib.pyplot as plt

folder_path = './confusion_matrices'

def plot_confusion_matrix(matrix, title):
    plt.figure(figsize=(4, 3))
    sns.heatmap(matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(title)
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.tight_layout()
    plt.show()

for filename in os.listdir(folder_path):
    if filename.endswith('.json'):
        json_path = os.path.join(folder_path, filename)

        with open(json_path, 'r') as f:
            confusion_matrices = json.load(f)

        for model, matrix in confusion_matrices.items():
            title = f"{filename} - {model}"
            print(f"Plotting confusion matrix for {title}")
            plot_confusion_matrix(matrix, title)
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

folder_path = './model_performance'

def plot_model_performance(csv_path, title):
    metrics_df = pd.read_csv(csv_path)

    # Plotting the metrics
    plt.figure(figsize=(10, 6))
    metrics_df.set_index('Model').plot(kind='bar', figsize=(10, 6), alpha=0.75)
    plt.title(title)
    plt.ylabel('Score')
    plt.xlabel('Model')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.legend(loc='lower right')
    plt.tight_layout()
    plt.show()

for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        csv_path = os.path.join(folder_path, filename)

        title = filename.replace('_', ' ').replace('.csv', '').capitalize()

        print(f"Plotting model performance for {title}")
        plot_model_performance(csv_path, title)
```
